```python
### import numpy as np
import cv2
from mss import mss
sct = mss()
from PIL import Image
import pyautogui as pg
import pydirectinput
import time
import math
import os.path


def target(target_filename):
    bounding_box = {'left': 0, 'top': 0, 'width': 1980, 'height': 1080}
    screen = sct.grab(bounding_box)
    source_img = np.array(screen)
    target_img = cv2.imread(target_filename, cv2.IMREAD_UNCHANGED)
    result = cv2.matchTemplate(source_img, target_img, cv2.TM_CCOEFF_NORMED)
    min_val, max_val, mix_loc, max_loc = cv2.minMaxLoc(result)
    w = target_img.shape[1]
    h = target_img.shape[0]

    threshold = .88

    yloc, xloc = np.where(result >= threshold)
    if(len(xloc) == 0):
        print('cannot detech')
        time.sleep(0.1)
        return target(target_filename)
    
    return xloc[0], yloc[0], w, h

def zhuhe():
    x,y,_,_ = target('zhuhe.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+10, y+120)
    pg.click(x+10, y+120)
    pydirectinput.moveTo(x+140, y+330)    
    pg.click(x+140, y+330)

    time.sleep(0.1)
    
def rearrange():
    x,y,_,_ = target('rearrange.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+10, y+10)
    pg.click(x+10, y+10)
    time.sleep(0.1)

def get_page():
    x,y,w,h = target('bag_page.png')
    if(x == 0) & (y == 0):
        return
    return x,y,w,h
    
def go_to_page(current_page, x, y, w, h):
    if (current_page > 4):
        return
    pydirectinput.moveTo(int(x+w/2), int(y+h/4*current_page-15))
    pg.click(int(x+w/2), int(y+h/4*current_page-15))
    print('Clicked page {0}'.format(current_page))
    time.sleep(0.1)
    
def search_and_build_item(filename):
    is_file_exists = os.path.exists(filename)
    if (is_file_exists != True):
        return
    
    bounding_box = {'left': 0, 'top': 0, 'width': 1980, 'height': 1080}
    target_img = cv2.imread(filename, cv2.IMREAD_UNCHANGED)
    count = 0
    while True:
        screen = sct.grab(bounding_box)
        source_img = np.array(screen)
        result = cv2.matchTemplate(source_img, target_img, cv2.TM_CCOEFF_NORMED)
        min_val, max_val, mix_loc, max_loc = cv2.minMaxLoc(result)
        w = target_img.shape[1]
        h = target_img.shape[0]

        threshold = .98

        yloc, xloc = np.where(result >= threshold)
        if(len(xloc) == 0):
            break

        x, y = math.floor(xloc[0]+w/2),math.floor(yloc[0]+h/2)

        pydirectinput.moveTo(x, y)
        pg.rightClick(x, y)
        pydirectinput.moveTo(x+20, y+50)
        pg.click(x+20, y+50)
        time.sleep(0.1)

        # zhu he
        zhuhe()
        count += 1
    return count
        
def refill(count):
    x,y,_,_ = target('refill_1.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+10, y+10)
    pg.click(x+10, y+10)
    time.sleep(0.1)
    
    x,y,_,_ = target('refill_2.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+10, y+10)
    pg.rightClick(x+10, y+10)
    time.sleep(0.1)
    
    x,y,_,_ = target('refill_3.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+10, y+10)
    pg.rightClick(x+10, y+10)
    time.sleep(0.1)
    
    x,y,_,_ = target('refill_4.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+10, y+10)
    pg.rightClick(x+10, y+10)
    time.sleep(0.1)
    
    x,y,w,_ = target('refill_5.png')
    if(x == 0) & (y == 0):
        return
    pydirectinput.moveTo(x+w-10, y+10)
    pg.click(x+w-10, y+10)
    time.sleep(0.1)
    
    for i in range(count):
        pydirectinput.typewrite('3')
        time.sleep(0.1)
    time.sleep(0.1)

def main():
    rearrange()

    page_x,page_y,page_w,page_h = get_page()
    
    count = 0
    for current_page in range(4):
        go_to_page(current_page+1, page_x,page_y,page_w,page_h)
        for level in range(4):
            for item in range(5):
                filename = 'item_{0}_{1}.png'.format(item, level)
                count += search_and_build_item(filename)
    print('done')
    print(count)

    refill(count)
    
main()
```
